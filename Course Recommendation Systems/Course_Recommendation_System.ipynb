```python
import pandas as pd
import numpy as np
import neattext.functions as nfx
import seaborn as sns

from sklearn.feature_extraction.text import TfidfVectorizer, CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity, linear_kernel
```

```python
df = pd.read_csv("/content/udemy_course_data.csv")
df.head()
```

```python
# Generating clean text by removing the stopwords and special characters

df['Clean_title'] = df['course_title'].apply(nfx.remove_stopwords)

df['Clean_title'] = df['Clean_title'].apply(nfx.remove_special_characters)

df['Clean_title'].iloc[1:5]
```

```python
# Vectorizing the course title

countvect = CountVectorizer()

cv_mat = countvect.fit_transform(df['Clean_title'])

cv_mat
```

```python
cv_mat.todense()
```

```python
df_cv_words = pd.DataFrame(
    cv_mat.todense(),
    columns = countvect.get_feature_names_out()
)


df_cv_words
```

```python
# Cosine similarity Matrix

cosine_sim_mat = cosine_similarity(cv_mat)

cosine_sim_mat
```

```python
def recommended_course(title, numrec =10):

  course_index = pd.Series(
      df.index, index = df['course_title']
  ).drop_duplicates()

  index = course_index[title]

  scores = list(enumerate(cosine_sim_mat[index]))

  sorted_scores = sorted(scores, key = lambda x:x[1], reverse = True)

  selected_course_index = [i[0] for i in sorted_scores[1:]]

  selected_course_score = [i[1] for i in sorted_scores[1:]]

  rec_df = df.iloc[selected_course_index]

  rec_df['Similarity_Score'] = selected_course_score

  final_recommended_courses = rec_df[[
      'course_title', 'Similarity_Score', 'url', 'price','num_subscribers'
  ]]

  return final_recommended_courses.head(numrec)
```

# Coding the GUI Implementation

```python
import pandas as pd
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity
import ipywidgets as widgets
from IPython.display import display
```

```python
# Function for recommending courses
def recommended_course(title, numrec=10):
    try:
        course_index = pd.Series(df.index, index=df['course_title']).drop_duplicates()
        index = course_index[title]

        scores = list(enumerate(cosine_sim_mat[index]))
        sorted_scores = sorted(scores, key=lambda x: x[1], reverse=True)

        selected_course_index = [i[0] for i in sorted_scores[1:]]
        selected_course_score = [i[1] for i in sorted_scores[1:]]

        rec_df = df.iloc[selected_course_index].copy()  # Create a copy of the DataFrame slice
        rec_df.loc[:, 'Similarity_Score'] = selected_course_score  # Use .loc[] to safely assign the new column

        final_recommended_courses = rec_df[['course_title', 'Similarity_Score', 'url', 'price', 'num_subscribers']]

        return final_recommended_courses.head(numrec)

    except KeyError:
        return pd.DataFrame()  # Return an empty DataFrame if the course is not found
```

```python
# Function to display recommendations in Colab
def on_button_click(b):
    course_title = course_title_widget.value
    num_recs = int(num_recs_widget.value) if num_recs_widget.value.isdigit() else 10

    recommendations = recommended_course(course_title, num_recs)

    if not recommendations.empty:
        output.clear_output()
        with output:
            for _, row in recommendations.iterrows():
                print(f"Course: {row['course_title']}")
                print(f"Similarity Score: {row['Similarity_Score']}")
                print(f"URL: {row['url']}")
                print(f"Price: {row['price']}")
                print(f"Subscribers: {row['num_subscribers']}")
                print("-" * 50)
    else:
        output.clear_output()
        with output:
            print("No recommendations found.")

# Create widgets for input
course_title_widget = widgets.Text(value='', placeholder='Enter Course Title', description='Course Title:', disabled=False)
num_recs_widget = widgets.Text(value='10', placeholder='Number of Recommendations', description='Number of Rec:', disabled=False)
button = widgets.Button(description="Get Recommendations")
output = widgets.Output()

# Link button click event to the function
button.on_click(on_button_click)

# Display the widgets
display(course_title_widget, num_recs_widget, button, output)
```
